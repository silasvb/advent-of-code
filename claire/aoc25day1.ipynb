```python
example_inputs = [
    "L68",
"L30",
"R48",
"L5",
"R60",
"L55",
"L1",
"L99",
"R14",
"L82"
]
```

```python
with open("../inputs/day1-claire.txt", "r") as f:
    example_inputs = [line.strip() for line in f.readlines()]
```

```python

position = 50

positions = []

count_of_zeros = 0

for input in example_inputs:    
    if input[0] == "L":
        distance = int(input[1:])
        position = position - distance
        
        if position < 0:
            position = 100 +    position
    
    else:
        distance = int(input[1:])
        position = position + distance
        
        if position >= 100:
            position = position - 100
    
    if position == 0:
        count_of_zeros = count_of_zeros + 1
        
    positions.append(position)
        
        
print("Number of zeros: " + str(count_of_zeros))
print("positions: " + str(positions))
```
