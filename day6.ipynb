```python
# Test Input

input = [
    "123 328  51 64 ",
    " 45 64  387 23 ",
    "  6 98  215 314",
    "*   +   *   +  "]
```

```python
# Read input

with open("inputs/day6.txt", "r") as f:
    input = f.read().splitlines()
```

```python
def parse_input(input: List[str]):
    
    operators = input[-1].split()
    problems = [(operator, []) for operator in operators]
    
    for line in input[:-1]:
        numbers = line.split()
        for i, number in enumerate(numbers):
            problems[i][1].append(number)
                
    return problems
    
problems = parse_input(input)  
print(problems)
```

```python
def create_equations(problems):
    equations = [operator.join(nums) for operator, nums in problems]
        
    return equations

equations = create_equations(problems)
print(equations)
```

```python
# execute
results = [eval(equation) for equation in equations]
print(results)

print("Part 1:", sum(results))
```
