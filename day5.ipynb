```python
from timer import timeit

# Generate example input
input = [
    "3-5",
"10-14",
"16-20",
"12-18",
"",
"1",
"5",
"8",
"11",
"17",
"32"
]
```

```python
def generate_ranges(input_lines):
    
    blank_idx = input_lines.index("")
    
    ranges = input_lines[:blank_idx]
    
    return [tuple(map(int, r.split("-"))) for r in ranges]


def generate_numbers(input_lines):
    blank_idx = input_lines.index("")
    
    numbers = input_lines[blank_idx + 1:]
    
    return [int(n) for n in numbers]


def sort_ranges(ranges):
    return sorted(ranges, key=lambda x: x[0])


numbers = generate_numbers(input)
sorted_ranges = sort_ranges(generate_ranges(input))
```

```python
@timeit
def solve(ranges, numbers):
    
    count = 0
    for num in numbers:
        if is_in_ranges(ranges, num):
            count += 1
    return count


def is_in_ranges(ranges, num):
    return any(start <= num <= end for start, end in ranges)
```

```python
@timeit
def solve_part_2(ranges):
    
    merged_ranges = []
    
    i = 0
    while i < len(ranges):
        current_start, current_end = ranges[i]
        
        j = i + 1
        while(ranges[j][0] <= current_end):
            current_end = max(current_end, ranges[j][1])
            j += 1
            if j >= len(ranges):
                break
        
        merged_ranges.append((current_start, current_end))
        
        i = j
        
    return sum(end - start + 1 for start, end in merged_ranges)
```

```python
# Solving with example input

result = solve(sorted_ranges, numbers)
print(result)

solve_part_2(sorted_ranges)
```
