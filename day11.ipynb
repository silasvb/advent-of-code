```python
input = """
aaa: you hhh
you: bbb ccc
bbb: ddd eee
ccc: ddd eee fff
ddd: ggg
eee: out
fff: out
ggg: out
hhh: ccc fff iii
iii: out
""".strip().splitlines()
```

```python
with open("inputs/day11.txt") as f:
    input = f.read().strip().splitlines()
```

```python
paths = {}
for line in input:
    node, outcomes = line.split(": ")
    paths[node] = outcomes.split(" ")


print(paths)

def find_paths(source, target, exclude):
    if source == target:
        return [[]]
    
    next = paths[source]
    all_paths = []

    for n in next:
        if n in exclude:
            continue
        
        sub_paths = find_paths(n, target, exclude + [source])
        for sp in sub_paths:
            all_paths.append([n] + sp)
        
    return all_paths
    
print(len(find_paths("you", "out", [])))
```

```python
def find_paths_cached(n, cache, fft_visited=False, dac_visited=False):

    if (n, fft_visited, dac_visited) in cache:
        return cache[(n, fft_visited, dac_visited)]

    # Handle Points of Interest
    if n == "fft":
        fft_visited = True
    if n == "dac": 
        dac_visited = True

    # Handle end terminal condition
    if n == "out":
        # Counts as an invalid path (and return 0) if we haven't visited both
        return fft_visited and dac_visited 
    
    result = sum(find_paths_cached(c, cache, fft_visited, dac_visited) for c in paths[n])
    cache[(n, fft_visited, dac_visited)] = result

    return result
    
print(find_paths_cached("svr", {}))
```
