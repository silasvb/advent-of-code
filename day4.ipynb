```python
# Parse input

with open('inputs/day4.txt') as f:
    lines = f.read().splitlines()
    

```

```python
lines = [
    "..@@.@@@@.",
"@@@.@.@.@@",
"@@@@@.@.@@",
"@.@@@@..@.",
"@@.@@@@.@@",
".@@@@@@@.@",
".@.@.@.@@@",
"@.@@@.@@@@",
".@@@@@@@@.",
"@.@.@@@.@."
]
```

```python
# Turn into matrix of 1s and 0s

matrix = [list(1 if c == '@' else 0 for c in line) for line in lines]

def print_matrix(mat):
    for row in mat:
        for c in row:
            print(c, end='')
        print("")
        
print_matrix(matrix)
```

```python
def can_ij_move(matrix, x, y, max_x, max_y, max_allowed_neighbours):
    l = 1 if x > 0 and matrix[y][x-1] == 1 else 0
    r = 1 if x < max_x and matrix[y][x+1] == 1 else 0
    u = 1 if y > 0 and matrix[y-1][x] == 1 else 0
    d = 1 if y < max_y and matrix[y+1][x] == 1 else 0
    ul = 1 if x > 0 and y > 0 and matrix[y-1][x-1] == 1 else 0
    ur = 1 if x < max_x and y > 0 and matrix[y-1][x+1] == 1 else 0
    dl = 1 if x > 0 and y < max_y and matrix[y+1][x-1] == 1 else 0
    dr = 1 if x < max_x and y < max_y and matrix[y+1][x+1] == 1 else 0
    
    return sum([l, r, u, d, ul, ur, dl, dr]) <= max_allowed_neighbours


def calculate_movable_positions(matrix):
    max_x = len(matrix[0]) - 1
    max_y = len(matrix) - 1

    movable_positions = []

    for y in range(len(matrix)):
        for x in range(len(matrix[0])):
            if matrix[y][x] == 1:
                can_move = can_ij_move(matrix, x, y, max_x, max_y, 3)
                if can_move:
                    movable_positions.append((x, y))
                
    return movable_positions

movable_positions = calculate_movable_positions(matrix)
print(len(movable_positions))
```

```python

for x, y in movable_positions:
    matrix[y][x] = 2
    
    
print_matrix(matrix)
```

```python
# Part 2

def update_matrix(matrix, movable_positions):
    for x, y in movable_positions:
        matrix[y][x] = 0
        
    return matrix


movable_positions = calculate_movable_positions(matrix)
print(f"{len(movable_positions)} movable rolls before updates")

iteration = 1
total_moved = len(movable_positions)

while(len(movable_positions) > 0):
    matrix = update_matrix(matrix, movable_positions)
    movable_positions = calculate_movable_positions(matrix)
    total_moved += len(movable_positions)
    print(f"Iteration {iteration}: {len(movable_positions)} more movable rolls - total moved: {total_moved}")
    iteration += 1
    
    
print(f"Total moved: {total_moved}")
```
