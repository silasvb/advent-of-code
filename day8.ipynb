```python
with open("inputs/day8.txt", "r") as f:
    input = f.readlines()

print(input)
total_joins = 1000
included_circuits = 3
```

```python
def preprocess(lines):
    for l in lines:
        vals = l.split(",")
        yield (int(vals[0]), int(vals[1]), int(vals[2]))
```

```python
import numpy

processed_input = list(preprocess(input))
arr = numpy.array(processed_input)

from scipy.spatial.distance import pdist, squareform

distances = squareform(pdist(arr))

min_distances = []
for i, row in enumerate(distances):
    for j in range(i + 1, len(row)):
        min_distances.append((row[j], (i, j)))

min_distances.sort()

joins = {c: set([c]) for c in range(len(arr))}

for _, (to, frm) in min_distances:
    
    combined = joins[to] | joins[frm]
    
    for k in combined:
        joins[k] = combined

    if len(combined) == len(joins):
        print("All joined")
        x_from = arr[frm][0]
        x_to = arr[to][0]

        print(f"x_from: {x_from}, x_to: {x_to} - {x_from * x_to}")

        break 


unique_sets = []
for k, vals in joins.items():
    if k == min(vals):
        unique_sets.append(vals)

print(unique_sets)
lengths = sorted([len(j) for j in unique_sets], reverse=True)

import math
print(f"Solution: {math.prod(lengths[0:included_circuits])}")
```
