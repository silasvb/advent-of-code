```python
input = [
    ".......S.......",
"...............",
".......^.......",
"...............",
"......^.^......",
"...............",
".....^.^.^.....",
"...............",
"....^.^...^....",
"...............",
"...^.^...^.^...",
"...............",
"..^...^.....^..",
"...............",
".^.^.^.^.^...^.",
"..............."
]
```

```python
with open('inputs/day7.txt') as f:
    input = [line.strip() for line in f.readlines()]
```

```python
def preprocess(input):
    start_idx = input[0].index('S')
    grid = [[c == '^' for c in line] for line in input[1:]]
    return start_idx, grid

start_idx, grid = preprocess(input)
print("Start Index:", start_idx)

current_verticals = {start_idx: 1}

splits = 0

for row in grid:
    new_verticals = {}
    for idx, val in current_verticals.items():
        if row[idx]:
            if idx == 0:
                if idx + 1 in new_verticals:
                    new_verticals[idx + 1] += val
                else:
                    new_verticals[idx + 1] = val
            elif idx == len(row) - 1:
                if idx - 1 in new_verticals:
                    new_verticals[idx - 1] += val
                else:
                    new_verticals[idx - 1] = val
                splits += 1
            else:
                if idx + 1 in new_verticals:
                    new_verticals[idx + 1] += val
                else:
                    new_verticals[idx + 1] = val

                if idx - 1 in new_verticals:
                    new_verticals[idx - 1] += val
                else:
                    new_verticals[idx - 1] = val
                splits += 1
        else:
            if idx in new_verticals:
                new_verticals[idx] += val
            else:
                new_verticals[idx] = val
    
    current_verticals = new_verticals
    print(" ".join(f"{current_verticals[idx]:4}" if idx in current_verticals else "   *" for idx in range(len(input[0]))))

print(f"Solution Part 1: {splits}")
print(f"Solution Part 2: {sum(v for _, v in current_verticals.items())}")


```
